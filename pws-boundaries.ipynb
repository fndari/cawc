```python
def select_good_geom(df):
    return (df
            # .dropna(subset=['geometry'])  # it's not necessary to do this as well,
            # it's already included in this condition
            .loc[lambda d: d.geom_type.astype(str).isin(["MultiPolygon"])]
           )
```

```python
PWS = PWS.pipe(select_good_geom)
PWS
```

Following the amazing resource at <https://source.opennews.org/articles/choosing-right-map-projection/>, and confirming at http://spatialreference.org/ref/epsg/nad83-california-albers/, I'm using `EPSG:3310` as the map projection for area calculations. 
To "calibrate" and check that the units are correct, I looked at Santa Clara Valley Water District. Its wikipedia page reports a coverage area of 3800 km^2, which matches with what we have after setting the projection.

```python
def calc_area(df, projection='epsg:3310'):
    conv_factor = 1e-6  # the area with this projection is in m^2, we should use km^2 or (ewwww) miles^2
    return (df
            .to_crs({'init': projection})
            .assign(area_km2=lambda d: d.geometry.area * conv_factor)
           )
```

```python
PWS = PWS.pipe(calc_area)
```

# Solve duplicates

Some entries have duplicates:

```python
PWS.groupby('pwsid').size().value_counts()
```

We can use this info to decide which duplicate to take:
- The `is_current` column
- The `notes` column contains comments mentioning whether the entry has been "retired" or "replaced"
- The various date columns do not seem to be very discriminating in this regard

```python
def has_duplicate(df):
    # the keep=False kwarg marks all entries as duplicate (as opposed to only the first or last encountered)
    return df.duplicated('pwsid', keep=False)

def mark_duplicates(df):
    
    return df.assign(has_duplicate=has_duplicate)

def is_retired(df):
    return (df.notes
            .str.contains('retired|replaced', case=False, na=False)
           )
```

```python
# PWS = PWS.astype({'is_current': bool})  # warning! this doesn't propagate the correct DataFrame subclass
PWS = PWS.assign(is_current=lambda d: d.is_current.astype(bool))
```
