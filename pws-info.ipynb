Now we standardize the data within each column. First we try to set categorical dtypes for all relevant columns, and then manually check the results for inconsistencies

```python
COLUMNS_CATEGORICAL =[
    'county_principal',
    'ws_fed_type_code',
    'ws_fed_type',
    'ws_state_type_code',
    'ws_state_type',
    'ws_status_code',
    'ws_status',
    'owner_type_code',
    'owner_type',
    'primary_source_type_code',
    'primary_source_type',
    'fee_code',
    'city',
    'treatment_plant_class_code',
    'treatment_plant_class',
    'distr_system_class_code',
    'distr_system_class']

def inspect_categories(df):
    for col in df:
        try:
            cat = df[col].cat.categories
        except AttributeError:
            pass
        else:
            print('-'*10)
            print(col)
            print(cat)
            
```

```python
PWS.astype({c: 'category' for c in COLUMNS_CATEGORICAL}).pipe(inspect_categories)
```

It seems that we need to:
- Fix inconsistency with whitespace
- `city`: this should probably be called "location", since we can already seen several non-municipality markers
    - Also, there are multiple entries referring to the same location, e.g. 'YOSEMITE', 'YOSEMITE NATIONAL PARK', 'YOSEMITE NTL PK'

```python
MISSING_VALUE_TEXT = 'nan'  # I don't know if this is a special value or not, but functions are already returning it

def get_consistent_labels(c):
    return (c.astype(str)
            .str.strip()
            .replace('', MISSING_VALUE_TEXT)
            .replace('nan', MISSING_VALUE_TEXT)
           )

def apply_assign(df, f, columns=None):
    "Apply the same function `f` to multiple `columns`, and assign the result in place"
    columns = columns or []
    # the double-starred syntax works because the dataframe is dict-like {key: pd.Series}
    # remember the (kind-of counterintuitive) axis correspondence:
    # axis=0 applies to each column, axis=1 applies to each row
    # in general, axis=0 -> OK, axis=1 -> much slower
    return df.assign(**(df[columns].apply(f, axis=0)))

def clean_categories(df, cols):
    return df.pipe(apply_assign, get_consistent_labels, cols)

def clean(df):
    return  (df
             .dropna(subset=['pws_id'])
             .pipe(clean_categories, COLUMNS_CATEGORICAL)
             .astype({c: 'category' for c in COLUMNS_CATEGORICAL})
            )
```

```python
(PWS
 .pipe(clean)
 .pipe(inspect_categories)
)
```

```python
SUBSTITUTIONS = {
    re.compile('^MT[.]*'): 'MT',
    (re.compile('^ST[.]*'), re.compile('^SAINT')): 'ST',
    ('NATL PK', 'NTL PK'): 'NATIONAL PARK',
    "SEQ NAT'L FOREST": 'SEQUOIA NATIONAL FOREST',
    'NATL': 'NATIONAL',
    'MTN. ': 'MOUNTAIN ',
    'MR. SHASTA': 'MT SHASTA',
    'AFB': 'AIR FORCE BASE',
}

VARIANTS = {
    ('MI WUK', 'MI-WUK'): 'MI-WUK',
}

MISSPELLINGS = {
    'MR. SHASTA': 'MT SHASTA',
    'HELENEA': 'HELENA',
    'FLINTRIGE': 'FLINTRIDGE',
}
```

```python
def uniformize_whitespace(c):
    return c.str.replace('(\s+)', ' ')

def remove_trailing_characters(c):
    return c.str.replace('([,.]*)$', '')

def match_unmatched_parens(c):
    return c.str.replace(r'[(](\w+)$', '(\g<1>)')

def apply_substitutions(c, subs=None):

    def adapt_substitutions(subs_in):
        "Allow both tuples and single values in input map, for convenience"
        def ensure_appropriate_sequence(item):
            print(f'item: {item}, type: {type(item)}')
            if not isinstance(item, tuple):
                item = tuple([item])
            print(f'item: {item}, type: {type(item)}')
            return item
        
        return {ensure_appropriate_sequence(patterns): target
                for patterns, target in subs_in.items()}

    subs = adapt_substitutions(subs)

    for patterns, target in subs.items():
        for pattern in patterns:
            c = c.str.replace(pattern, target)
    return c

def apply_all_substitutions(c):
    for subs in [MISSPELLINGS, SUBSTITUTIONS, VARIANTS]:
        c = apply_substitutions(c, subs)
    return c
```

```python
OPERATIONS = [
    lambda c: c,
    uniformize_whitespace,
    compose(remove_trailing_characters, uniformize_whitespace),
    compose(match_unmatched_parens, remove_trailing_characters, uniformize_whitespace),
    compose(apply_all_substitutions,
            match_unmatched_parens, remove_trailing_characters, uniformize_whitespace),
]

def inspect_n_categories(c):
    c = c.astype('category')
    print(f'{c.name}: {len(c.cat.categories)}')

def compare_effect(original, operations, tester=inspect_n_categories):
    for op in operations:
        tester(op(original))
```

```python
compare_effect(PWS.city, OPERATIONS)
```

```python
def uniformize_location(df):
    operations = [lambda c: c.astype('category'),
                  apply_all_substitutions,
                  match_unmatched_parens,
                  remove_trailing_characters,
                  uniformize_whitespace]
    # works as intended now, after fixing a bug in `apply_assign`
    return df.pipe(apply_assign, compose(*operations), columns=['city'])
#     return df.assign(city=lambda d: compose(*operations)(d.city))
```
